```python
# For numerical arrays
import numpy as np

# Matplotlib for plotting
import matplotlib.pyplot as plt
plt.close('all')
```

```python
data = np.genfromtxt('results/bx0p5_data.csv', delimiter=',')
expData = np.genfromtxt('results/Hua1986_bx0p5_disp.csv', delimiter=',')
```

```python
# Set plot font to size 14
font = {'size'   : 14}
plt.rc('font', **font)

# Get array of default plot colors
prop_cycle = plt.rcParams['axes.prop_cycle']
colors     = prop_cycle.by_key()['color']

# Only plot as far as we have time history data
ind = np.argmax(data[0,:]) +1

plt.figure()
# plt.axvline(0.0017, c='k', linewidth=1.)
plt.axhline(0, c='k', linewidth=1.)
plt.plot(data[0,0:ind], -data[1,0:ind], linewidth=2.0, color=colors[3], label="Simulation")#, marker='.')
plt.scatter(expData[:,0] - expData[0,0], expData[:,1] - expData[0,1], s=25,
                     edgecolors= (0.0, 0.0, 0.0,1),
                     color=(0,0,0,1),
                     linewidth=1.0, label="Experiment")
plt.axis('tight')
plt.xlabel(r"Time (s)")
plt.ylabel(r"Displacement (mm)")
plt.grid(linestyle="--", linewidth=0.5, color='b')
# plt.ylim(-6,8)
plt.xlim(0,0.15)
plt.legend()

fig = plt.gcf()
fig.set_size_inches(7,5)
plt.tight_layout()
plt.savefig("results/tip_disp_curve_lorentz.png", dpi=600)
```
